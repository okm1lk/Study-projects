```python
# Количество книг, выпущенных после 01.01.2001
query = '''
SELECT count(book_id) as count_books
FROM books
WHERE  publication_date > '2000-01-01'
'''
```

```python
get_sql_data(query)
```

```python
# Название, количество обзоров и средняя оценка книг
query = '''
SELECT a.title, count_reviews, avg_rating
FROM (
      SELECT title, count(review_id) as count_reviews
      FROM books b JOIN reviews rv
      ON b.book_id = rv.book_id
      GROUP BY title) a
JOIN (
      SELECT title, avg(rating) as avg_rating
      FROM books b JOIN ratings rg
      ON b.book_id=rg.book_id
      GROUP BY title) b
ON a.title=b.title
ORDER BY count_reviews DESC, avg_rating DESC
'''
```

```python
get_sql_data(query)
```

```python
# Издательство, выпустившее больше всего книнг толще 50 страниц
query = '''
SELECT publisher, count(book_id) as count_books
FROM publishers p JOIN books b
ON p.publisher_id = b.publisher_id
WHERE num_pages > 50
GROUP BY publisher
ORDER BY count_books DESC
LIMIT 1
'''
```

```python
get_sql_data(query)
```

```python
# Автор с самым высоким средним рейтингом, имеющий 50 и более оценок
query = '''
SELECT author, round(avg(rating), 2) as avg_rating, count(rating_id) as count_scores
FROM authors a JOIN books b
ON a.author_id = b.author_id
JOIN ratings r
ON b.book_id = r.book_id
GROUP BY author
HAVING count(rating_id) >= 50
ORDER BY avg_rating DESC
LIMIT 1
'''
```

```python
get_sql_data(query)
```

```python
# Среднее количество обзоров от пользователей, которые поставили более 48 оценок
query = '''
SELECT avg(count_reviews_by_user) as avg_reviews
FROM (
      SELECT count(DISTINCT review_id) as count_reviews_by_user
      FROM reviews rv JOIN ratings rg
      ON rv.username = rg.username
      GROUP BY rg.username
      HAVING count(DISTINCT rating_id) > 48
      ) a
'''
```

```python
get_sql_data(query)
```
