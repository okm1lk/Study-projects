## Общая информация

### Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. 

Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.  

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## Предобработка данных

### Изменение названий столбцов

```python
df.columns
```

```python
df = df.rename(columns={'Name':'name', 'Platform':'platform', 'Year_of_Release':'release_year', 'Genre':'genre', \
                        'NA_sales':'na_sales','EU_sales':'eu_sales', 'JP_sales':'jp_sales','Other_sales':'other_sales', \
                        'Critic_Score':'critic_score', 'User_Score':'user_score', 'Rating':'rating'})
df.columns
```

Все названия столбцов преведены к snake_case, а столбец `Year_Of_Release` переименован в `release_year`

### Обработка пропусков

```python
df.isna().sum()
```

```python
# определяю % пропусков в каждом столбце
round((df.isna().sum() / df.isna().count())  * 100, 2)
```

В столбцах `name` и `genre` менее 0.1% пропусков, можно сразу удалить эти строки  
В столбце `release_year` 1.6 % пропусков, также можно удалить

Сначала проверю, нет ли в этих строках значимых продаж

```python
df[(df['name'].isna() == True) | df['genre'].isna() == True]
```

```python
df[(df['release_year'].isna() == True) & ((df['na_sales'] > 1) | (df['eu_sales'] > 1) | (df['jp_sales'] > 1))]
```

Очень высоких показателей не обнаружил, можно удалять

```python
df = df.dropna(subset=['name', 'genre', 'release_year'])
df.isna().sum()
```

```python
# т.к. пропусков в этом столбце больше нет, изменяю тип данных на int
df['release_year'] = df['release_year'].astype(int)
```

Рассмотрим отдельно столбцы `user_score`, `critic_score` и `rating`

```python
df['user_score'].unique()
```

В солбце есть строки с рейтингом TBD, это значит, что рейтинг пока не определен, но появится в будущем. Заменю эти значения на пропуски, чтобы изменить тип данных столбца, но не удалять эти строки  

```python
df ['user_score'] = df['user_score'].replace({'tbd':np.nan})
df['user_score'] = pd.to_numeric(df['user_score'])

df['user_score'].dtype
```

```python
df['critic_score'].unique()
# оставляем все как есть, заменить пропуски нечем
```

```python
df['rating'].unique()
```

По информации из [Википедии](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board)

*«E» («Everyone») — «Для всех»*: Содержание вполне подходит для всех возрастов. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первоначально *"K-A" ("Kids to Adults")*  

**Значения `K-A` можно заменить на `E`, т.к. это одно и то же**  

Пропуски заменим значением `unrated`


```python
df['rating'] = df['rating'].replace({'K-A': 'E'})
df['rating'] = df['rating'].fillna('unrated')
df['rating'].unique()
```

### Добавление новых столбцов

Добавляю столбец с суммарными продажами по всем регионам

```python
df['all_sales'] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
df.head(5)
```

### Поиск и удаление дубликатов

```python
df.duplicated().sum()
```

```python
df[df.duplicated(subset=['name', 'platform', 'release_year']) == True]
```

```python
df[(df['name'] == 'Madden NFL 13') & (df['platform'] == 'PS3')]
```

Нашелся один неявный дубликат, удаляем

```python
df = df.drop_duplicates(subset=['name', 'platform', 'genre', 'release_year'])
df[(df['name'] == 'Madden NFL 13') & (df['platform'] == 'PS3')]
```

**Вывод**  

- Все названия столбцов преведены к нижнему регистру, а столбец `Year_Of_Release` переименован в `release_year`  
  
- В столбцах `name` и `genre` менее 0.1% пропусков, поэтому эти строки были удалены  
  
- В столбце `release_year` 1.6% пропусков, эти строки также были удалены, а тип данных столбца изменен га `int`
  
- В столбце `user_score` есть строки с рейтингом TBD, это значит, что рейтинг пока не определен, но появится в будущем. Заменю эти значения на пропуски, чтобы изменить тип данных. Тип данных изменил на `float`  
  
- В столбце `rating` заменил значение 'K-A' на 'E', т.к. это один и тот же возрастной рейтинг, чтобы уменьшить количество категорий
  
- Явных дубликатов не обнаружено, но нашелся один неявный и был удален

- Добавлен новый столбец `all_sales` с суммарными продажами по всем регионам

### Срок жизни платформ

```python
# создаю список из 10 платтформ с наибольшими суммарными продажами

platform_list = (
    df.groupby('platform')['all_sales'].sum()
    .sort_values(ascending=False).head(10)
    .reset_index()['platform'].tolist()
)
platform_list
```

```python
platform_lifetime_plot = (
    df
    .query('platform in @platform_list and release_year >= 1994')
    .pivot_table(index='release_year', values='all_sales', columns='platform', aggfunc='sum')
)
```

```python
year_ticks = df[df['release_year'] >= 1994]['release_year'].unique().tolist() 

fig, ax = plt.subplots(figsize=(12,6))

ax = sns.lineplot(data=platform_lifetime_plot, dashes=False)

ax.set_title('Продажи по платформам и годам')
ax.set(xlabel='Год', ylabel='Продажи, млн. копий')
ax.legend(title='Платформа')
ax.set_xticks(year_ticks)
ax.set_xticklabels(year_ticks, rotation=45)
plt.show()
```

По этому графику видно, что обычно платформы живут около 10 лет, поэтому актуальным можно считать период 3-4 года и рассматривать платформы, по которым в это время росли продажи, т.к. они еще не достигли своего пика. Будем считать актуальным период 2013-2016 гг

### Влияние отзывов на продажи

Рассмотрим влияние отзывов внутри платформы X360

```python
df_360 = df_actual.query('platform == "X360"')

print(df_360[['critic_score', 'user_score', 'all_sales']].corr())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax1 = sns.scatterplot(ax=axes[0], data=df_360, x='all_sales', y='user_score', alpha=0.7)
ax1.set_title('Зависимость продаж от оценок пользователей для X360')
ax1.set(ylabel='Оценка пользоваталей', xlabel='Продажи, млн. копий')

ax2 = sns.scatterplot(ax=axes[1], data=df_360, x='all_sales', y='critic_score', alpha=0.7)
ax2.set_title('Зависимость продаж от оценок критиков для X360')
ax2.set(ylabel='Оценка критиков', xlabel='Продажи, млн. копий')

plt.show()
```

Оценки пользователей практически никак не влияют на продажи игр  

Оценки критиков, незначительно, но влияют продажи. Скорее всего свзяано с тем, что критики получают игру до ее релиза и заранее выставляют оценку, на которую некоторые пользователи могут ориентироваться и принять решение о покупке

Проверим это предположение на других платформах

```python
platform_list = ['PS3', 'X360', 'PS4', '3DS', 'XOne', 'PC']

for platform in platform_list:
    df_actual_platform = df_actual.query('platform == @platform')
    print('Платформа', platform)
    print(df_actual_platform[['critic_score', 'user_score', 'all_sales']].corr())

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    ax1 = sns.scatterplot(ax=axes[0], data=df_actual_platform, x='all_sales', y='user_score', alpha=0.7)
    ax1.set_title(f'Зависимость продаж от оценок пользователей для {platform}')
    ax1.set(ylabel='Оценка пользоваталей', xlabel='Продажи, млн. копий')

    ax2 = sns.scatterplot(ax=axes[1], data=df_actual_platform, x='all_sales', y='critic_score', alpha=0.7)
    ax2.set_title(f'Зависимость продаж от оценок критиков для {platform}')
    ax2.set(ylabel='Оценка критиков', xlabel='Продажи, млн. копий')

    plt.show()
```

## Проверка гипотез

### Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H_0: Средние пользовательские рейтинги платформы Xbox One = Средние пользовательские рейтинги платформы PC  
H_1: Средние пользовательские рейтинги платформы Xbox One ≠ Средние пользовательские рейтинги платформы PC

```python
xone = df_actual.query('platform == "XOne"')['user_score'].dropna().tolist()
pc = df_actual.query('platform == "PC"')['user_score'].dropna().tolist()

results = st.ttest_ind(xone, pc)

alpha = 0.05

print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Гипотеза 2: Средние пользовательские рейтинги жанров Action и Sports  разные.

H_0: Средние пользовательские рейтинги жанров Action = Средние пользовательские рейтинги жанров Sports  
H_1: Средние пользовательские рейтинги жанров Action ≠ Средние пользовательские рейтинги жанров Sports

```python
action = df.query('genre == "Action"')['user_score'].dropna().tolist()
sports = df.query('genre == "Sports"')['user_score'].dropna().tolist()

results = st.ttest_ind(action, sports)

alpha = 0.05

print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
